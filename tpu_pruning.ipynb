```python
import onnx
from onnx_tf.backend import prepare
```

```python
try:
    tpu = tf.distribute.cluster_resolver.TPUClusterResolver()
    if tpu.cluster_spec() is not None:
        print('Running on TPU')
        tf.config.experimental_connect_to_cluster(tpu)
        tf.tpu.experimental.initialize_tpu_system(tpu)
        strategy = tf.distribute.experimental.TPUStrategy(tpu)
    else:
        print('No TPU available')
except ValueError:
    print('TPU not found')

# Now `strategy` is defined, so you can use it here
with strategy.scope():  # This ensures that the model is built on the TPU
    tf_model = tf_rep.export_graph("/content/drive/MyDrive/models/base/20/tf_model")
```
